```python
import numpy as np
import matplotlib.pyplot as plt
from timeit import default_timer as timer

def insertionSort(list):
    for index in range(1, len(list)):
        temp = list[index]
        pos = index - 1
        while pos >= 0 and temp < list[pos]:
            list[pos + 1] = list[pos]
            pos -= 1
        pos += 1
        list[pos] = temp
x_values = [10000, 20000, 50000, 100000]
random_times = [0] * len(x_values)
ascending_times = [0] * len(x_values)
descending_times = [0] * len(x_values)
for i in range(len(x_values)):
    randomNumbers = np.random.randint(1, 10000, x_values[i])
    descendingNumbers = list(range(x_values[i], 0, -1))
    totalTime = 0
    repetitions = 5
    for trials in range(repetitions):
        start_time = timer()
        insertionSort(randomNumbers)
        end_time = timer()
        totalTime += end_time - start_time
    totalTime /= repetitions
    random_times[i] = totalTime * 1000
    #
    totalTime = 0
    for trials in range(repetitions):
        start_time = timer()
        insertionSort(descendingNumbers)
        end_time = timer()
        totalTime += end_time - start_time
    totalTime /= repetitions
    descending_times[i] = totalTime * 1000

plt.plot(x_values, random_times, label = "Random")
plt.plot(x_values, descending_times, label = "Descending")
#insertionSort(y_values)

#plt.plot(x_values, y_values, label = "Random")#descending, ascending

plt.xlabel("Number of Items Sorted")
plt.ylabel("Time (milliseconds)")
plt.title("Insertion Sort")
plt.legend()
plt.show()
```

```python
def selectionSort(list):
    for i in range(0, len(list) - 1):
        minIndex = i
        for j in range(i + 1, len(list)):
            if list[j] < list[minIndex]:
                minIndex = j
        temp = list[minIndex]
        list[minIndex] = list[i]
        list[i] = temp

x_values = [10000, 20000, 50000, 100000]
random_times = [0] * len(x_values)
ascending_times = [0] * len(x_values)
descending_times = [0] * len(x_values)
for i in range(len(x_values)):
    randomNumbers = np.random.randint(1, 10000, x_values[i])
    descendingNumbers = list(range(x_values[i], 0, -1))
    totalTime = 0
    repetitions = 5
    for trials in range(repetitions):
        start_time = timer()
        selectionSort(randomNumbers)
        end_time = timer()
        totalTime += end_time - start_time
    totalTime /= repetitions
    random_times[i] = totalTime * 1000
    #
    totalTime = 0
    for trials in range(repetitions):
        start_time = timer()
        selectionSort(descendingNumbers)
        end_time = timer()
        totalTime += end_time - start_time
    totalTime /= repetitions
    descending_times[i] = totalTime * 1000

plt.plot(x_values, random_times, label = "Random")
plt.plot(x_values, descending_times, label = "Descending")
#insertionSort(y_values)

#plt.plot(x_values, y_values, label = "Random")#descending, ascending

plt.xlabel("Number of Items Sorted")
plt.ylabel("Time (milliseconds)")
plt.title("Selection Sort")
plt.legend()
plt.show()
```

```python
def merge(left, right):
    output = []
    leftIndex = 0
    rightIndex = 0
    while leftIndex < len(left) and rightIndex < len(right):
        if left[leftIndex] < right[rightIndex]:
            output.append(left[leftIndex])
            leftIndex += 1
        else:
            output.append(right[rightIndex])
            rightIndex += 1
    if leftIndex == len(left):
        output.extend(right[rightIndex:])
    else:
        output.extend(left[leftIndex:])
    return output

def mergeSort(list):
    if len(list) <= 1:
        return list
    left = mergeSort(list[:len(list) // 2])
    right = mergeSort(list[len(list) // 2:])
    return merge(left, right)
            
x_values = [10000, 20000, 50000, 100000]
random_times = [0] * len(x_values)
ascending_times = [0] * len(x_values)
descending_times = [0] * len(x_values)
for i in range(len(x_values)):
    randomNumbers = np.random.randint(1, 10000, x_values[i])
    descendingNumbers = list(range(x_values[i], 0, -1))
    totalTime = 0
    repetitions = 5
    for trials in range(repetitions):
        start_time = timer()
        randomNumbers = mergeSort(randomNumbers)
        end_time = timer()
        totalTime += end_time - start_time
    totalTime /= repetitions
    random_times[i] = totalTime * 1000
    #
    totalTime = 0
    for trials in range(repetitions):
        start_time = timer()
        descendingNumbers = mergeSort(descendingNumbers)
        end_time = timer()
        totalTime += end_time - start_time
    totalTime /= repetitions
    descending_times[i] = totalTime * 1000

plt.plot(x_values, random_times, label = "Random")
plt.plot(x_values, descending_times, label = "Descending")
#insertionSort(y_values)

#plt.plot(x_values, y_values, label = "Random")#descending, ascending

plt.xlabel("Number of Items Sorted")
plt.ylabel("Time (milliseconds)")
plt.title("Merge Sort")
plt.legend()
plt.show()
```
